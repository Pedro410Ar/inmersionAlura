El Instituto Nacional Central Único Coordinador de Ablación e Implante impulsa, normatiza, coordina y fiscaliza las actividades de donación y trasplante de órganos, tejidos y células en nuestro país.
https://www.argentina.gob.ar/salud/incucai

El Sistema Nacional de Información de Procuración y Trasplante de la República Argentina (SINTRA) es implementado por el INCUCAI. Dicho sistema informático permite la administración, gestión, fiscalización y consulta de la actividad de procuración y trasplante de órganos, tejidos y células en el ámbito nacional.
https://sintra.incucai.gov.ar/

El SINTRA administra el Registro Nacional de Donantes de Órganos y Tejidos (RENADON), que tiene como objetivo primario instrumentar el registro único de ciudadanos que expresaron su voluntad hacia la donación de órganos y/o tejidos en el ámbito nacional.
https://sintra.incucai.gov.ar/donar/

En el Registro puede manifestar su voluntad en forma gratuita cualquier argentino mayor de edad, por vía digital, personal o por correo postal.
image.png

La importancia del Registro consiste justamente en la presunción que establece la Ley argentina: si no hay manifestación expresa en contra, se presume que la persona cuando fallece es donante.
Por ello, la Base de Datos del Registro está estructurada con columnas donde hay una Manifestación Positiva y una Manifestación Negativa. Ambas pueden realizarse.
